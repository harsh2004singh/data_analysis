# apply OneHotEncoding on Zone Column

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one = OneHotEncoder()
```

```python
encoded = one.fit_transform(data[['Zone']]).toarray()
```

```python
df = pd.DataFrame(encoded, columns=one.get_feature_names_out(['Zone']))
```

```python
df
```

```python
data=pd.concat([data,df],axis=1)
```

```python
data.head(2)
```

```python
# We working Zones Not in state data so 
data.drop(columns =['State','Zone'],inplace = True)
```

# Label Encoding on Occupation

```python
data['Occupation'] = data['Occupation'].str.strip()
Occ_count = data['Occupation'].value_counts()
```

```python
Occ_count
```

```python
less_500 = Occ_count[Occ_count<=500]
less_500
```

```python
data['Occupation']=data['Occupation'].apply(lambda x: "other" if x in less_500 else x)
```

```python
data['Occupation'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data['Occ_encod'] = le.fit_transform(data['Occupation'])
```

```python
data['Prod_encode']= le.fit_transform(data['Product_Category'])
```
